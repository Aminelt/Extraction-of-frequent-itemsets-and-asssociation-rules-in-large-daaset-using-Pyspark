```python
for x in df.index:
  if df.loc[x, "BMI"] < 18.4:
    df.loc[x, "BMI"] = 'UnderWeight'
  elif( df.loc[x, "BMI"] >=18.5) &  (df.loc[x, "BMI"] <=24.9 ) :
        df.loc[x, "BMI"] = 'Normal weight'
  elif   ( df.loc[x, "BMI"] >=25) &  (df.loc[x, "BMI"] <=29.9): 
    df.loc[x, "BMI"] = ' OverWeight'
  elif   ( df.loc[x, "BMI"] >=30) & (df.loc[x, "BMI"] <=34.9): 
    df.loc[x, "BMI"] = ' Obeisity class I'
  elif   ( df.loc[x, "BMI"] >=35.0) & ( df.loc[x, "BMI"] <=39.9): 
    df.loc[x, "BMI"] = ' Obeisity class II'
  elif    df.loc[x, "BMI"] >=40 : 
    df.loc[x, "BMI"] = ' Obeisity class III'
```
