Model **Building**

```python
diabetes_copy = diabetes_data.copy()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
diabetes_data = diabetes_data .apply(le.fit_transform)
diabetes_data .head()
```

```python
# Importing Libraries
!pip install KModes
from kmodes.kmodes import KModes
```

**Searching for K** 

```python
cost = []
for num_clusters in list(range(1,5)):
    kmode = KModes(n_clusters=num_clusters, init = "Cao", n_init = 1, verbose=1)
    kmode.fit_predict(diabetes_data)
    cost.append(kmode.cost_)
```

```python
y = np.array([i for i in range(1,5,1)])
plt.plot(y,cost)
```

```python
km_cao = KModes(n_clusters=3, init = "Cao", n_init = 1, verbose=1)
fitClusters_cao = km_cao.fit_predict(diabetes_data)
```

```python
fitClusters_cao
```

```python
diabetes_data = diabetes_copy.reset_index()
```

```python
clustersDf = pd.DataFrame(fitClusters_cao)
clustersDf.columns = ['cluster_predicted']
combinedDf = pd.concat([diabetes_data, clustersDf], axis = 1).reset_index()
combinedDf = combinedDf.drop(['index', 'level_0'], axis = 1)
```
